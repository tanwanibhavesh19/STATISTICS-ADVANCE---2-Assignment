1. What is hypothesis testing in statistics ?
  - Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data.
  - It involves testing an assumption (called a hypothesis) about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?
  - The null hypothesis assumes no effect or relationship between variables, while the alternative hypothesis proposes a relationship or effect exists.

3. What is the significance level in hypothesis testing, and why is it important ?
  - In hypothesis testing, the significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It essentially defines the threshold of evidence required to reject the null hypothesis. Why is it important? It controls the risk of making a Type I error (false positive), ensuring the validity and reliability of the findings.

4. What does a P-value represent in hypothesis testing ?
  - In hypothesis testing, a p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. It helps determine the strength of evidence against the null hypothesis, guiding decisions about whether to reject it or not.

5. How do you interpret the P-value in hypothesis testing ?
  - A small p-value (typically less than or equal to 0.05) suggests strong evidence that the observed results are unlikely to have occurred by chance if the null hypothesis were true, leading to its rejection. Conversely, a large p-value (greater than 0.05) suggests insufficient evidence to reject the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing ?
  - In hypothesis testing, Type 1 and Type 2 errors are two types of incorrect conclusions that can be drawn when testing a null hypothesis. A Type 1 error occurs when a true null hypothesis is rejected, while a Type 2 error occurs when a false null hypothesis is not rejected.

8. What is the Z-test, and when is it used in hypothesis testing ?
  - A z-test is a hypothesis testing method used to determine if there's a statistically significant difference between two population means or between a sample mean and a population mean when the population standard deviation is known and the sample size is large (generally 30 or greater). It's based on the normal distribution and provides a way to assess the probability of observing a sample mean or difference in means if the null hypothesis were true.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing ?
  - A z-score represents how many standard deviations a data point is from the mean of a distribution. It's calculated by subtracting the mean from the data point and dividing the result by the standard deviation. In hypothesis testing, the z-score helps determine if the observed data is significantly different from what would be expected under the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution ?
  - The t-distribution is a probability distribution similar to the normal distribution but with heavier tails, meaning it has a greater chance of extreme values. It's used as an alternative to the normal distribution when dealing with small sample sizes (n<30) and when the population standard deviation (σ) is not known.

11. What is the difference between a Z-test and a T-test ?
  -  The main difference between a z-test and a t-test lies in the assumption about the population variance. A z-test requires the population variance to be known, or the sample size to be large (n > 30). T-tests, on the other hand, are used when the population variance is unknown, or when the sample size is small (n < 30).

12. What is the T-test, and how is it used in hypothesis testing ?
  - A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups. It's commonly used to compare the means of two groups or to compare the mean of a sample to a known value.

13. What is the relationship between Z-test and T-test in hypothesis testing ?
  - In hypothesis testing, Z-tests and t-tests are both used to compare sample means to population means or to compare means between two groups, but they differ in their assumptions and how they handle the population standard deviation.

14. What is a confidence interval, and how is it used to interpret statistical results ?
  - A confidence interval is a range of values used to estimate an unknown population parameter based on sample data. It provides a level of confidence (e.g., 95%) that the true population parameter falls within that range. Confidence intervals are crucial for interpreting statistical results because they quantify the uncertainty in estimates and help assess the precision of the findings.

15. What is the margin of error, and how does it affect the confidence interval ?
  - The margin of error (ME) indicates how much a sample statistic (like a survey result) might differ from the true population parameter. It's essentially half the width of a confidence interval. A smaller ME means the estimate is more precise, while a larger ME suggests greater uncertainty. The ME directly impacts the confidence interval, affecting its width and therefore the level of confidence in the estimate.

16. How is Bayes' Theorem used in statistics, and what is its significance ?
  - Bayes' Theorem is a fundamental formula in probability and statistics used to calculate conditional probability. It allows you to update your belief about an event based on new evidence, incorporating prior knowledge with observed data to arrive at a revised or posterior probability.

17. What is the Chi-square distribution, and when is it used ?
  - The chi-square distribution is a continuous probability distribution used in hypothesis testing, particularly for analyzing categorical data. It assesses whether the observed frequencies of a variable's categories differ significantly from the expected frequencies based on the proposed distribution. The test statistic (Chi-Square value) is calculated by comparing observed and expected frequencies, and a p-value is determined to decide if the difference is significant, indicating a poor fit.

19. What is the F-distribution, and when is it used in hypothesis testing ?

It's commonly used in the F-test, which is a statistical test to determine if there are significant differences in the means of two or more groups. The F-distribution helps determine the probability of observing a given F-statistic under the assumption that there are no real differences between the groups being compared.

It analyzes the variance within and between groups to see if the difference between the means is more significant than the variation within the groups. ANOVA's validity relies on certain assumptions being met.

```python
#2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters of the discrete uniform distribution
low = 1       # inclusive lower bound
high = 7      # exclusive upper bound (so this simulates 1 to 6)

# Create a discrete uniform distribution object
dist = randint(low, high)

# Define the support (possible values)
x = np.arange(low, high)

# Compute PMF values
pmf_values = dist.pmf(x)

# Plot the PMF
plt.figure(figsize=(8, 5))
plt.stem(x, pmf_values, use_line_collection=True)
plt.title('PMF of Discrete Uniform Distribution (Die Roll)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
        x (int): The value (0 or 1)
        p (float): The probability of success (0 ≤ p ≤ 1)

    Returns:
        float: The probability P(X = x)
    """
    if x not in [0, 1]:
        return 0.0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p if x == 1 else (1 - p)

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Output: 0.7
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Output: 0.3
print("P(X=2):", bernoulli_pdf(2, 0.7))  # Output: 0.0 (invalid x)


```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a given dataset and plots them.

    Parameters:
        data (array-like): The input dataset (list or NumPy array)
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Compute Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-scores')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='red', linestyle='--', linewidth=1, label='±1σ')
    plt.axhline(-1, color='red', linestyle='--', linewidth=1)
    plt.title('Z-Scores of Dataset')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=30)
z = calculate_and_plot_z_scores(sample_data)
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the original non-normal distribution
lambda_param = 1  # Rate parameter for exponential distribution

# CLT parameters
sample_size = 30         # Number of observations per sample
num_samples = 10000      # Number of samples

# Step 1: Generate sample means from the exponential distribution
sample_means = []

for _ in range(num_samples):
    sample = np.random.exponential(scale=1/lambda_param, size=sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 2: Plot histogram of sample means
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Sample Means')

# Step 3: Overlay the normal distribution with mean and std of sample means
mean_clt = np.mean(sample_means)
std_clt = np.std(sample_means)
x = np.linspace(min(bins), max(bins), 1000)
plt.plot(x, norm.pdf(x, mean_clt, std_clt), 'r-', lw=2, label='Normal Approximation')

plt.title(f'Central Limit Theorem Demonstration\nSample size = {sample_size}, Number of samples = {num_samples}')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, point):
    """
    Calculate the Z-score of a given data point.

    Parameters:
        data (array-like): Dataset to compute mean and std.
        point (float): Data point to calculate the Z-score for.

    Returns:
        float: Z-score of the data point.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z = (point - mean) / std
    return z

def plot_z_score(z):
    """
    Plot the standard normal distribution and mark the Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label='Standard Normal PDF')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='skyblue', alpha=0.5, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
dataset = [55, 60, 65, 70, 75, 80, 85, 90, 95]
data_point = 80

z_score = calculate_z_score(dataset, data_point)
print(f"Z-score for data point {data_point}: {z_score:.2f}")

plot_z_score(z_score)
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for the sample mean.

    Parameters:
        sample_data (array-like): Sample observations.
        population_mean (float): Hypothesized population mean under H0.
        population_std (float): Known population standard deviation.
        alpha (float): Significance level (default 0.05).

    Returns:
        dict: Contains z_statistic, p_value, and test conclusion.
    """
    sample = np.array(sample_data)
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    std_error = population_std / np.sqrt(n)

    # Compute Z-statistic
    z_stat = (sample_mean - population_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    reject_null = p_value < alpha
    conclusion = "Reject the null hypothesis." if reject_null else "Fail to reject the null hypothesis."

    return {
        "z_statistic": z_stat,
        "p_value": p_value,
        "conclusion": conclusion
    }

# Example usage:
sample_data = [52, 55, 47, 51, 49, 50, 53, 48, 54, 46]  # Sample observations
population_mean = 50                                   # Hypothesized population mean
population_std = 5                                     # Known population std deviation
alpha = 0.05                                          # Significance level

result = z_test(sample_data, population_mean, population_std, alpha)
print(f"Z-statistic: {result['z_statistic']:.3f}")
print(f"P-value: {result['p_value']:.4f}")
print(result['conclusion'])

```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default 0.95 for 95% CI)

    Returns:
        tuple: (lower_bound, upper_bound) of the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean

    # t critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
sample_data = [55, 60, 65, 70, 75, 80, 85, 90, 95]
confidence_level = 0.95

lower, upper = confidence_interval(sample_data, confidence=confidence_level)
mean_val = np.mean(sample_data)

print(f"Sample mean = {mean_val:.2f}")
print(f"{confidence_level*100:.0f}% confidence interval: ({lower:.2f}, {upper:.2f})")
print(f"Interpretation: We are {confidence_level*100:.0f}% confident that the true population mean lies within this interval.")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean_true = 100
std_true = 15
sample_size = 50

sample_data = np.random.normal(loc=mean_true, scale=std_true, size=sample_size)

# Step 2: Calculate the confidence interval for the mean
def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return mean, lower_bound, upper_bound

mean_sample, lower_ci, upper_ci = confidence_interval(sample_data)

# Step 3: Print and interpret results
print(f"Sample Mean: {mean_sample:.2f}")
print(f"95% Confidence Interval for the Mean: ({lower_ci:.2f}, {upper_ci:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies within this interval.")

# Optional: show how close it is to the true mean
print(f"True Mean was: {mean_true}")
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test comparing sample mean to population mean.

    Parameters:
        sample_data (array-like): Sample observations.
        pop_mean (float): Known population mean.
        pop_std (float): Known population standard deviation.
        alpha (float): Significance level (default 0.05).

    Returns:
        dict: Contains z_statistic, p_value, and conclusion.
    """
    sample = np.array(sample_data)
    n = len(sample)
    sample_mean = np.mean(sample)

    # Standard error of the mean
    std_error = pop_std / np.sqrt(n)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: Sample mean is significantly different from population mean."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference from population mean."

    return {
        "z_statistic": z_stat,
        "p_value": p_value,
        "conclusion": conclusion
    }

# Example usage:
np.random.seed(1)
sample = np.random.normal(loc=52, scale=10, size=30)  # Sample data
population_mean = 50                                 # Known population mean
population_std = 12                                  # Known population std deviation
alpha_level = 0.05                                  # Significance level

result = z_test(sample, population_mean, population_std, alpha_level)
print(f"Z-statistic: {result['z_statistic']:.3f}")
print(f"P-value: {result['p_value']:.4f}")
print(result['conclusion'])
```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_error = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical z values for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

    # Shade acceptance region
    plt.fill_between(x, 0, y, where=( (x > -z_critical) & (x < z_critical) ), color='green', alpha=0.3, label='Acceptance Region')

    # Mark observed z_stat
    plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Observed Z = {z_stat:.2f}')

    plt.title(f'Two-tailed Z-test (α={alpha})\nZ-statistic = {z_stat:.3f}, p-value = {p_value:.4f}')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    # Conclusion
    if abs(z_stat) > z_critical:
        conclusion = "Reject the null hypothesis."
    else:
        conclusion = "Fail to reject the null hypothesis."

    return z_stat, p_value, conclusion

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=40)
population_mean = 50
population_std = 12
alpha_level = 0.05

z_stat, p_val, concl = two_tailed_z_test(sample_data, population_mean, population_std, alpha_level)
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(concl)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean_null, pop_mean_alt, pop_std, n, alpha=0.05):
    # Standard error
    se = pop_std / np.sqrt(n)

    # Critical z values for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Critical sample means (boundaries for rejection region)
    crit_low = pop_mean_null - z_crit * se
    crit_high = pop_mean_null + z_crit * se

    # Define x range for plotting
    x_min = min(pop_mean_null, pop_mean_alt) - 4*se
    x_max = max(pop_mean_null, pop_mean_alt) + 4*se
    x = np.linspace(x_min, x_max, 1000)

    # PDFs under null and alternative hypotheses
    pdf_null = norm.pdf(x, loc=pop_mean_null, scale=se)
    pdf_alt = norm.pdf(x, loc=pop_mean_alt, scale=se)

    plt.figure(figsize=(12, 6))

    # Plot null hypothesis distribution
    plt.plot(x, pdf_null, label='H0 Distribution (mean={})'.format(pop_mean_null), color='black')
    # Plot alternative hypothesis distribution
    plt.plot(x, pdf_alt, label='H1 Distribution (mean={})'.format(pop_mean_alt), color='green')

    # Shade rejection regions under H0 (Type I error regions)
    plt.fill_between(x, 0, pdf_null, where=(x <= crit_low), color='red', alpha=0.4, label='Type I Error Region (α)')
    plt.fill_between(x, 0, pdf_null, where=(x >= crit_high), color='red', alpha=0.4)

    # Shade acceptance region under H0 (where we fail to reject)
    plt.fill_between(x, 0, pdf_null, where=((x > crit_low) & (x < crit_high)), color='grey', alpha=0.2, label='Acceptance Region')

    # Shade Type II error region under H1 (β)
    plt.fill_between(x, 0, pdf_alt, where=((x > crit_low) & (x < crit_high)), color='blue', alpha=0.4, label='Type II Error Region (β)')

    # Mark critical values
    plt.axvline(crit_low, color='red', linestyle='--')
    plt.axvline(crit_high, color='red', linestyle='--')

    plt.title('Type I and Type II Errors in Two-Tailed Z-Test')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    # Calculate beta (Type II error probability)
    beta = norm.cdf(crit_high, loc=pop_mean_alt, scale=se) - norm.cdf(crit_low, loc=pop_mean_alt, scale=se)

    print(f"Significance level (Type I error, α): {alpha}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage:
plot_type1_type2_errors(
    pop_mean_null=100,
    pop_mean_alt=105,
    pop_std=15,
    n=30,
    alpha=0.05
)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    """
    Perform an independent two-sample t-test and interpret results.

    Parameters:
        group1 (array-like): Sample data for group 1.
        group2 (array-like): Sample data for group 2.
        alpha (float): Significance level (default 0.05).

    Returns:
        dict: t_statistic, p_value, conclusion
    """
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch’s t-test

    if p_value < alpha:
        conclusion = f"Reject the null hypothesis (p = {p_value:.4f} < {alpha}). The means are significantly different."
    else:
        conclusion = f"Fail to reject the null hypothesis (p = {p_value:.4f} ≥ {alpha}). No significant difference between means."

    return {
        "t_statistic": t_stat,
        "p_value": p_value,
        "conclusion": conclusion
    }

# Example usage:
np.random.seed(0)
sample_group1 = np.random.normal(loc=50, scale=8, size=35)
sample_group2 = np.random.normal(loc=54, scale=8, size=40)

result = independent_t_test(sample_group1, sample_group2)
print(f"T-statistic: {result['t_statistic']:.4f}")
print(f"P-value: {result['p_value']:.4f}")
print(result['conclusion'])
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
        sample (array-like): Sample data.
        confidence (float): Confidence level (default 0.95 for 95%).

    Returns:
        tuple: (lower_bound, upper_bound)
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # sample std dev, ddof=1 for unbiased estimator

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=50)

ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval for the mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
        sample (array-like): Sample data.
        confidence (float): Confidence level (default 0.95).

    Returns:
        float: Margin of error.
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * std_err
    return moe

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

moe_95 = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe_95:.3f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): P(H), prior probability of hypothesis.
        likelihood (float): P(E|H), probability of evidence given hypothesis.
        evidence (float): P(E), total probability of evidence.

    Returns:
        float: Posterior probability P(H|E).
    """
    return (likelihood * prior) / evidence

# Given data
P_disease = 0.01           # Prior P(D)
P_no_disease = 1 - P_disease

P_pos_given_disease = 0.99 # Sensitivity P(Pos|D)
P_neg_given_no_disease = 0.95 # Specificity P(Neg|¬D)
P_pos_given_no_disease = 1 - P_neg_given_no_disease

# Calculate total probability of positive test (evidence)
P_pos = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Calculate posterior probability P(Disease | Positive test)
posterior = bayes_theorem(P_disease, P_pos_given_disease, P_pos)

print(f"Probability of having disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
from scipy.stats import chi2_contingency

# Example data:
# Suppose we surveyed people's favorite pet type across two age groups

# Contingency table:
# Rows = Age groups: Young, Old
# Columns = Favorite pet: Dog, Cat, Fish
observed = np.array([
    [50, 30, 20],  # Young
    [30, 50, 40]   # Old
])

# Perform Chi-square test for independence
chi2, p_value, dof, expected = chi2_contingency(observed)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p_value:.4f}")
print("Expected frequencies if independent:")
print(np.round(expected, 2))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variables are dependent (there is an association).")
else:
    print("Fail to reject the null hypothesis: No evidence of association (variables are independent).")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test given observed data.

    Parameters:
        observed (2D array-like): Observed contingency table.

    Returns:
        numpy.ndarray: Expected frequencies table.
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    expected = row_sums @ col_sums / total
    return expected

# Example usage:
observed_data = [
    [30, 10, 20],
    [20, 20, 40]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Observed frequencies:")
print(np.array(observed_data))
print("\nExpected frequencies:")
print(np.round(expected_freq, 2))
```

```python
#14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Observed frequencies from rolling a dice 60 times (example)
observed = np.array([8, 9, 10, 11, 12, 10])

# Expected frequencies assuming a fair dice (equal probability)
expected = np.full_like(observed, fill_value=np.sum(observed)/len(observed))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: The observed data fits the expected distribution.")
```
